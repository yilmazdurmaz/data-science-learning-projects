```python
import pandas as pd
```

```python
df = pd.read_csv("https://earthquake.usgs.gov/fdsnws/event/1/query.csv?starttime=1900-01-01%2000:00:00&minmagnitude=7&orderby=time")
```

```python
data=df[["time","place","mag"]].sort_values("time",ascending=False)
```
